# Plotly and Cufflinks

Plotly is a library that allows you to create interactive plots that you can use in dashboards or websites (you can save them as html files or static images).

## Installation

In order for this all to work, you'll need to install plotly and cufflinks to call plots directly off of a pandas dataframe. These libraries are not currently available through **conda** but are available through **pip**. Install the libraries at your command line/terminal using:

    pip install plotly
    pip install cufflinks

** NOTE: Make sure you only have one installation of Python on your computer when you do this, otherwise the installation may not work. **

## Imports and Set-up

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

print(__version__) # requires version >= 1.9.0
```

```python
import cufflinks as cf
```

```python
# For Notebooks
init_notebook_mode(connected=True)
```

```python
# For offline use
cf.go_offline()
```

### Fake Data

```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
```

```python
df2.head()
```

## Using Cufflinks and iplot()

* scatter
* bar
* box
* spread
* ratio
* heatmap
* surface
* histogram
* bubble

## Scatter

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=10)
```

```python
import plotly.express as px

px.scatter(df, x='A', y='B')
# fig.show()
```

## Bar Plots

```python
# 
import plotly.express as px

fig = px.bar(df2, x='Category', y='Values')
fig.show()
```

```python
# df.count().iplot(kind='bar')
px.bar(df.count().reset_index(), x='index', y=df.count().name)
```

## Boxplots

```python
# # df.iplot(kind='box')
# import plotly.express as px

# fig = px.box(df)
# fig.show()
px.box(df)

```

## 3d Surface

```python
# df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
# df3.iplot(kind='surface',colorscale='rdylbu')
import plotly.graph_objects as go
go.Figure(go.Surface(z=df2.values, colorscale='RdYlBu')).show()
go.figure(go.Surface(z=df2.values, colorscale='')).show()
```

## histogram

```python
# 
# df['A'].iplot(kind='hist',bins=25)
px.histogram(df, x='A', nbins=25, title='Histogram of A')
px.histogram(df, x='A', nbins=25, color_discrete_sequence=['skyblue'], title='Histogram of A')

```

```python
# df.iplot(kind='bubble',x='A',y='B',size='C')
import plotly.express as px

px.scatter(df, x='A', y='B', size='C', size_max=30, title='Bubble Plot')
```

## scatter_matrix()
