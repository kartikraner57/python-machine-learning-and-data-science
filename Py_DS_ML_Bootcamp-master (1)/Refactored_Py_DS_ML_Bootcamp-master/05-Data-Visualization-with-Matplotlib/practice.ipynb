```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
# FUNCTIONAL
plt.plot(x,y,'r-')
# plt.show()
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#00
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])

ax1.plot(x,y)
ax2.plot(y,x)

ax1.set_title('Larger Plot')
ax2.set_title('Smaller Plot')
```

```python
fig = plt.figure(figsize=(10,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize = (8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
# part 3
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y**2,label='X vs Y Squared')
ax.plot(y,x**4,label='Y vs X to the 4th power')
ax.set_title('My Title')
ax.set_xlabel('X Label')    
ax.set_ylabel('Y Label')
ax.legend(loc=0,)
```
