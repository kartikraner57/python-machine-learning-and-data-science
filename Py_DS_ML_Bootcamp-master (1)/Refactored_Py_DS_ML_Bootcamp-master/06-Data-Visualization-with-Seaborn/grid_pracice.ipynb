```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris.head() 
```

```python
iris['species'].unique(
```

```python
sns.pairplot(iris,)
```

```python
sns.pairplot(iris,hue='species',palette='Set2')
```

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')   
```

```python
g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip'  )
```

```python
# regression plot
sns.lmplot(x='total_bill', y='tip', data=tips,markers=['o','v'],scatter_kws={'s':100}, hue='smoker', palette='Set1')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', palette='Set1', col='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.6,)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex',markers=['o','v'], scatter_kws={'s':100})
```
