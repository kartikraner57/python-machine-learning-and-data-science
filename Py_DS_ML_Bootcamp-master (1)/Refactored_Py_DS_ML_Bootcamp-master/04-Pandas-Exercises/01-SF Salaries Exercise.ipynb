** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
def chief_count(title):
    if 'chief' in title.lower().split():
        return True
    else:
        return False
```

```python
chief_count('CHIEF MANAGER - METROPOLITAN TRANSIT AUTHORITY')
```

```python
# sf['JobTitle'].apply(lambda x: chief_count(x)).sum()
sum(sf['JobTitle'].apply(lambda x: chief_count(x)))
```
